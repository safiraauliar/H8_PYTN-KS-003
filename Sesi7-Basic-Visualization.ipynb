Question : create a box plot to visualize the distribution of the top 15 coutries (based on total immigration)
groupby the decades 1980s, 1990s,2000s

```python
# step 1: get dataset
df_top15 = df_can.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

step 2: membuat dataframe baru yang berisi agregat untuk setiap dekade, salah satu cara untuk mendapatkannya:
- create a list of all years in decade's 80s, 90s, 00s.
- slice the original dataframe df_can to create a series for each decade and sum across all years for each country.
- Merge the three series into a new dataframe. call your dataframe new_df

```python
years_80s = list(map(str, range(1980,1990)))
years_90s = list(map(str, range(1990,2000)))
years_00s = list(map(str, range(2000,2010)))

df_80s = df_top15.loc[:, years_80s].sum(axis=1)
df_90s = df_top15.loc[:, years_90s].sum(axis=1)
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

new_df = pd.DataFrame({'1980s':df_80s, '1990s':df_90s, '2000s':df_00s})
new_df.head()
```

### Bubble Plots
variasi dari scatter plot yang menampilkan 3 dimensi (x,y,z)

menganalisa dampak krisis Argentina dengan membandingkannya dengan negara Brazil terhadap jumlah penduduk yang bermigrasi ke Canada,

```python
# step 1: get data from Argentina and Brazil. convert years to type int and bring it in the dataframe
df_can_t = df_can[years].transpose()
#cast the Years (the index) to type int
df_can_t.index = map(int, df_can_t.index)

#let's label the index. this will automatically by the column name when we reset the index
df_can_t.index.name = 'Year'
#reset index to bring the Year in as a column
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
#Create Normalized weights

#Normalize Brazil Data
norm_brazil = (df_can_t['Brazil']-df_can_t['Brazil'].min())/(df_can_t['Brazil'].max()-df_can_t['Brazil'].min())

# Normalize Argentina Data
norm_argentina = (df_can_t['Argentina']-df_can_t['Brazil'].min())/(df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

Step 3: plot data
- to plot two different scatter plot, we can include the axes one plot into the other by passing it via the ax parameter 
- we will also pass in the weight using the s parameter. Given that the normalized weights are between 0-1, they won't be visible on the plot. therefore will be:
        -multiply weights by 2000 to scale it up on the graph, and
        -add 10 to compensate for the min value (which has a 0 weight and therefore scale with x2000)

```python
# Brazil
ax0= df_can_t.plot(kind='scatter',
                  x='Year',
                  figsize=(14,8),
                  alpha= 0.5,
                  color='green',
                  s=norm_brazil * 2000 +10,
                  xlim= (1975,2015),
                  y= 'Brazil')

#Argentina
ax1= df_can_t.plot(kind='scatter',
                  x='Year',
                  y='Argentina',
                  figsize=(14,8),
                  alpha= 0.5,
                  color="blue",
                  s= norm_argentina * 2000 +10,
                  ax=ax0
                  )
ax0.set_title("Immigration from Brazil and Argentina from 1980 - 2013")
ax0.set_ylabel("Number of Immigration")
ax0.legend(['Brazil', 'Argentina'], loc= "upper left", fontsize= 'x-large')
```

Question : Create bubble plots of immigration from China and India to visualize any differences with time from
1980 to 2013.you can use df_can_t that we defined and used in the previous example.

```python
# step 1: normalize data pertaining to China and India 
norm_china = (df_can_t['China'] - df_can_t['China'].min())/(df_can_t['China'].max() - df_can_t['China'].min())
norm_india = (df_can_t['India'] - df_can_t['India'].min())/(df_can_t['India'].max() - df_can_t['India'].min())
```
