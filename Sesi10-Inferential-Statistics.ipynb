```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns

from scipy.stats import uniform
```

```python
n=10000
start= 10
width= 20

dat_uniform= uniform.rvs(size=n, loc=start, scale=width)
dat_uniform
```

```python
data_plot= sns.distplot(dat_uniform,bins=10, color='red')
data_plot.set(xlabel='uniform Dis', ylabel= 'frequency')
```

```python
data_plot= sns.displot(dat_uniform,bins=10, color='red')
data_plot.set(xlabel='uniform Dis', ylabel= 'frequency')
```

```python
# normal distribusion
from scipy.stats import norm
data_norm= norm.rvs(size=n, loc=start, scale=width)
norm_plot= sns.distplot(data_norm, bins=20, color='blue', rug=True)
```

```python
# gamma distribution chi-squared
from scipy.stats import gamma

dat_gamma= gamma.rvs(size=n, a=5)
dat_gamma= sns.distplot(dat_gamma, bins=50, color='blue')

dat_gamma= gamma.rvs(size=n, a=5)
dat_gamma= sns.displot(dat_gamma, bins=50, color='blue')
```

```python
# exponential
from scipy.stats import expon

dat_expon= expon.rvs(size=n, loc=0, scale=1)
dat_gamma= sns.distplot(dat_expon, bins=50, color='blue')

dat_expon= expon.rvs(size=n, loc=0, scale=1)
dat_gamma= sns.displot(dat_expon, bins=50, color='blue')
```

```python
#poisson distribution
from scipy.stats import poisson
dat_poisson= poisson.rvs(size=n, mu=5)
dat_poisson= sns.distplot(dat_poisson, bins=50, color='blue')

```

```python
dat_poisson= poisson.rvs(size=n, mu=5)
dat_poisson= sns.displot(dat_poisson, bins=50, color='blue')
```
