```python
df_can.drop(['Coverage', 'AREA', 'REG', 'DEV'],axis=1, inplace=True)
```

```python
df_can.rename({'OdName':'Country'},axis=1, inplace=True)
df_can['Total'] = df_can.sum(axis=1)
df_can
```

```python
imigran_by_country = df_can.drop(['Type','AreaName','RegName','DevName'],axis =1)
imigran_by_country
```

```python
simigran_by_country=imigran_by_country.set_index('Country')
```

```python
#matplotlib  visualization
import matplotlib.pyplot as plt


AfghanistanImigran=simigran_by_country.loc['Afghanistan',range(1980,2013)]
AfghanistanImigran.plot()
plt.title('Imigran from Afghanistam')
plt.ylabel('Number of imigran')
plt.xlabel('Years')
plt.show()
```

```python
years = range(1980,2013)
```

```python
India_china_imigran = simigran_by_country.loc[['India','China'],years]
```

```python
India_china_imigrans = India_china_imigran.transpose()
India_china_imigrans.plot()
plt.title('Imigran from India and China')
plt.ylabel('Number of imigran')
plt.xlabel('Years')
plt.show()
```
