### Other Example of Confidence Intervals

```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import numpy as np
import random 
import warnings
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline

sns.set(rc={'figure.figsize':(13, 7.5)})
sns.set_context('talk')
```

```python
# Generate data of people's mass in pounds

np.random.seed(42)
normal_distribution_us_male_mass_pounds = np.random.normal(loc=181, scale=24, size=6000)
normal_distribution_us_female_mass_pounds = np.random.normal(loc=132, scale=22, size=6500)
all_mass_values = np.concatenate((normal_distribution_us_male_mass_pounds,
                                  normal_distribution_us_female_mass_pounds), axis=0)
df_ppl_mass= pd.DataFrame(data={'us_people_mass_pounds': all_mass_values})
```

```python
df_ppl_mass.head()
```

```python
# VIEW DISTRIBUTION OF US PEOPLE'S MASS

sns.distplot(df_ppl_mass['us_people_mass_pounds'], color='darkslategrey')
plt.xlabel("mass [pound]", labelpad=14)
plt.ylabel('Probability of Occurence', labelpad=14)
plt.title("Distribution Of Mass People in US", y=1.015, fontsize=20)
```

```python
# Calculation Population Mean
pop_mean_mass= df_ppl_mass['us_people_mass_pounds'].mean()
pop_mean_mass
```

```python
# Calculate population standard deviation
pop_std_mass = df_ppl_mass['us_people_mass_pounds'].std()
pop_std_mass
```

```python
sample_means = []
n=25
for i in range(0, 300):
    # random sample done with replacement
    sample_values=np.random.choice(a=df_ppl_mass['us_people_mass_pounds'], size= n)
    sample_mean= np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
# View Distribution of sample means (otherwise called sample distribution)
sns.distplot(sample_means)
plt.title("Distribution of Sample Means ($n=$25) of People's Mass in Pounds", y=1.015, fontsize=20)
plt.xlabel("sample mean mass [pounds]", labelpad=14)
plt.ylabel("frequency of occurence", labelpad=14)
```

#### Calculate Sampling Distribution (n=25) summary statistics
Berikut ini adalah ringkasan statistik untuk distribusi pengambilan sample yang divisualisasikan diatas

```python
# Calculate Median of sample mean
median_of_sample_means= np.median(sample_means)
median_of_sample_means
```

```python
# Calculate mean of sample mean
mean_of_sample_means = np.mean(sample_means)
mean_of_sample_means
```

```python
pop_mean_mass
```

```python
# calculate standard deviation of sample means
std_of_sample_means = np.std(sample_means)
std_of_sample_means
```

#### Equation for Standard Deviation of Sampling Distribution 
standard deviation dari sample means lebih sering disebut standard error (SE). Persamaan untuk SE :
* SE=/n
*  is population standard deviation
* n is sample size

```python
standard_error = pop_std_mass/np.sqrt(n)
standard_error
```

```python
std_of_sample_means
```

Nilai standard error ini sama dengan nilai yang dihitung di atas untuk std_of_sample_means.

#### Critical Z-score
kira-kira 95% nilai akan berada dalam 2 standard errors dari mean.Kita tahu bahwa ekor pada setiap ujung harus 2,5% atau 0,025 dari total area under the curve dari sampling distribution.

* Di bawah ini kita menggunakan z-score 1,96 untuk menghitung ekor 0,025.

```python
tail = 1 - stats.norm.cdf(1.96)
tail
```

```python
# the area under the curve between the z-critical  score is approximately 0.95
1-(tail*2)
```

```python
# menghitung batas untuk area dimana 95% mean sample berada di mean populasi
lower_95_perc_bound= mean_of_sample_means -1.96*standard_error
lower_95_perc_bound
```

```python
upper_95_perc_bound= mean_of_sample_means + 1.96*standard_error
upper_95_perc_bound
```

```python
kde= stats.gaussian_kde(sample_means)
pos= np.linspace(np.min(sample_means), np.max(sample_means), 10000)
plt.plot(pos, kde(pos), color='teal')
shade= np.linspace(lower_95_perc_bound, upper_95_perc_bound, 300)
plt.fill_between(shade, kde(shade), alpha=0.45, color='teal')
plt.text(x=154, y=0.01, horizontalalignment= 'center', fontsize=17,
        s="95% of sample means fall within\nthis shaded area of plus or minus \n1.96 z-score for the mean",
        bbox= dict(facecolor='whitesmoke', boxstyle='round, pad=0.1'))
plt.title("Distribution of sample means ($n=25$) of people's mass in pounds", y=1.015, fontsize=20)
plt.xlabel("sample means mass [pounds]", labelpad=14)
plt.ylabel("frequency of occurence", labelpad=14)
```

#### Hypothesis Testing
hypothesis testing dalam statistik adalah cara kita menguji hasil survei atau eksperimen untuk melihat apakah kita memiliki hasil yang bermakna. 

##### Research Question :
Pada tahun-tahun sebelumnya, 52% orang tua percaya bahwa elektronik dan media sosial adalah penyebab kurang tidur remaja mereka. Apakah saat ini semakin banyak orang tua yang percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik dan media sosial?
* Population: Orang tua dengan remaja (usia 13-18)
* Parameter of Interest: p
* Null Hypothesis: p = 0.52
* Alternative Hypthosis: p > 0.52 (note that this is a one-sided test)

Data: 1018 orang disurvei. 56% dari mereka yang disurvei percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik dan media sosial.

```python
n= 1018
pnull=0.52
phat= 0.56

sm.stats.proportions_ztest(phat*n, n, pnull, alternative='larger')
```

##### another example of hypothesis test

```python
url = "https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv"
da= pd.read_csv(url)
da.head()
```

```python
females= da[da['RIAGENDR']==2]
males= da[da['RIAGENDR']==1]
```

```python
n1= len(females)
mu1= females['BMXBMI'].mean()
sd1= females['BMXBMI'].std()

n1, mu1, sd1
```

```python
n2= len(males)
mu2= males['BMXBMI'].mean()
sd2= males['BMXBMI'].std()

n2, mu2, sd2
```

```python
sm.stats.ztest(females['BMXBMI'].dropna(), males['BMXBMI'].dropna(), alternative='two-sided')
```

##### Conclusion of the hypothesis test
Karena p-value (6.59e-10) sangat kecil, kita dapat  reject the Null hypothesis  bahwa rata-rata BMI laki-laki sama dengan perempuan. Catatan, kita menggunakan alternative="two-sided" dalam z-test karena di sini kita memeriksa ketidaksamaan.
Kita juga dapat memplot histogram dari data untuk memeriksa apakah kira-kira distribusinya mengikuti distribusi Normal.

```python
plt.figure(figsize=(7,4))
plt.title("Female BMI histogram", fontsize=16)
plt.hist(females['BMXBMI'].dropna(), edgecolor='k', color='pink', bins=25)
plt.show()

plt.figure(figsize=(7,4))
plt.title("Male BMI histogram", fontsize=16)
plt.hist(males['BMXBMI'].dropna(), edgecolor='k', color='blue', bins= 25)
plt.show()
```

### Hypothesis Testing for One Proportion

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist

df= pd.read_csv('./heart.csv')
df.head()
```

###### Question :
* The population proportion of Ireland having heart disease is 42%. Are more people suffering from heart disease in the US?

###### Step 1 : Define null hypothesis and alternative hypothesis
* null hypothesis -> population proportion yang menderita penyakit jantung di AS <= 42%
* alternative hypothesis -> proporsi penduduk amerika serikat yang menderita penyakit jantung < 42%

##### Step 2 : 
Assume that the dataset above is representative of sample from the population of the US. so calculate the population proportion of the US having heart disease.

```python
p_us= len(df[df['target']==1])/len(df)
p_us
```

population proportion sample yang mengalami penyakit jantung adalah 0.51 atau 51% presentase ini lebih besar dari null hypothesis yakni 42%

##### Step 3 : Calculate the test statistic
Rumus test statistic : (Best estimate - Hypothesized Estimate)/standard error of estimate
